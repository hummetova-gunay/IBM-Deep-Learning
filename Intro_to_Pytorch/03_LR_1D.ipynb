```python
import torch as t
```

```python
# y = -1 +2x

w = t.tensor(2.0, requires_grad=True)
b = t.tensor(-1.0, requires_grad=True)
```

```python
def forward(x):
    y_hat = b+ w*x
    return y_hat
```

```python
# Predict y = 2x - 1 at x = 1

x = t.tensor([[1.0]])
y_hat = forward(x)
print(y_hat)
```

```python
# Create x Tensor and check the shape of x tensor

x = t.tensor([[1.0], [2.0]])
print("The shape of x: ", x.shape)
```

```python
# Make the prediction of y = 2x - 1 at x = [1, 2]

yhat = forward(x)
print("The prediction: ", yhat)
```

```python
#  Class Linear
from torch.nn import Linear
```

```python
t.manual_seed(42)
```

```python
lr = Linear(in_features=1, out_features=1, bias = True)
print(list(lr.parameters()))
```

```python
# state_dict
print("Python dictionary: ",lr.state_dict())
print("keys: ",lr.state_dict().keys())
print("values: ",lr.state_dict().values())
```

```python
print("weight:",lr.weight)
print("bias:",lr.bias)
```

```python
#  make a single prediction at x = 1.0
x = t.tensor([[1.0]])
y_hat = lr(x)
print(y_hat)
```

```python
#  create multiple predictions
x = t.tensor([[1.0], [2.0]])
y_hat = lr(x)
print(y_hat)
```

```python
# custom modules
from torch import nn
```

```python
class LR(nn.Module):
    def __init__(self, input_size, output_size):
        # inherit from parent
        super(LR, self).__init__()
        self.linear = nn.Linear(input_size, output_size)

    def forward(self, x):
        out = self.linear(x)
        return out


```

```python
lr = LR(1,1)
print(list(lr.parameters()))
print(lr.linear)
```

```python
x = t.tensor([[1.0]])
y_hat = lr(x)
print(y_hat)
```

```python
x = t.tensor([[1.0], [2.0]])
yhat = lr(x)
print("The prediction: ", yhat)
```
