```python
import pandas as pd
import numpy as np
import keras
```

```python
filepath='https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv'
concrete_data = pd.read_csv(filepath)

concrete_data.head()
```

```python
concrete_data.shape
```

```python
concrete_data.isna().any()
```

```python
predictors = concrete_data[concrete_data.columns[concrete_data.columns!='Strength']]
predictors.head()
```

```python
target = concrete_data['Strength']
target.head()
```

```python
predictors_norm = (predictors - predictors.mean())/predictors.std()
predictors_norm.head()
```

```python
n_cols = predictors_norm.shape[1]
n_cols
```

### Fit a regression model with Keras


```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Input
```

```python
# build a neural network
def regression_model():
    model = Sequential()
    model.add(Input(shape=(n_cols,)))
    model.add(Dense(50, activation='relu'))
    model.add(Dense(50, activation='relu'))
    model.add(Dense(1))

    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
# build the model
model = regression_model()
```

```python
# fit the model
model.fit(predictors_norm, target, validation_split=.3, epochs=100, verbose=2)

# verbose decides how much training information is printed to the console.
```
